```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# importing seaborn so we can use the datasets
import seaborn as sns

%matplotlib inline
```

```python
df = sns.load_dataset("tips")

print(df.info())

df.head()
```

## Scatter Plot

```python
df.plot(x='total_bill', y='tip', kind='scatter', title='Pandas Scatter Plot Demo');
```

## Bar plot


```python
df.groupby('day').agg('sum').reset_index().plot(x='day', y='total_bill', kind='bar')
```

## Histogram

```python
df.hist(column='total_bill')
```

## Manipulating the plotting area with matplotlib

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 9))

df.plot(x='total_bill', y='tip', kind='scatter', ax=ax1)
df.plot(y='total_bill', kind='kde', ax=ax2)
```

## Plotting Functions in Pandas

## Lag Plots

http://pandas.pydata.org/pandas-docs/stable/visualization.html#visualization-lag

```python
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))

fig, ax = plt.subplots(figsize=(12, 8))

pd.plotting.lag_plot(data, ax=ax)
```

## Autocorrelation plots

http://pandas.pydata.org/pandas-docs/stable/visualization.html#autocorrelation-plot

```python
fig, ax = plt.subplots(figsize=(12, 8))

data = pd.Series(0.7 * np.random.rand(1000) + 0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))

pd.plotting.autocorrelation_plot(data, ax=ax)
```
